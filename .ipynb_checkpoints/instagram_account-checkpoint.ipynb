```python
import instaloader
import pandas as pd
import matplotlib.pyplot as plt
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
nltk.download('punkt')
nltk.download('stopwords')
from textblob import TextBlob
import spacy
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import string


# Initialize Instaloader
loader = instaloader.Instaloader()
```

## get captions from Instagram
we define a function with 2 parameters:
- account: pass the instagram account you want to get its caption
- number: tell how many post from the latest posts of the account you willing to get (default=50)

```python
def get_captions(account, number=50):
    captions = []
    profile = instaloader.Profile.from_username(loader.context, account)
    for post in profile.get_posts():
        if post.caption is not None:
            captions.append(post.caption)
            if len(captions) >=number:
                break
    return captions
```

## count the number of each word in all the recieved posts caption
this function is for counting the number of words in the captions <br>
remove parameter in this function is a list of arbitrary word that you dont want to be counted:
- for example : link in bio , is a typical sentence in any instagram post. we can remove these words for better analysing

```python
 def count_words(captions, remove=None):
    all_words = []
    for caption in captions:
        words = word_tokenize(caption.lower())
        all_words.extend(words)
    stop_words = set(stopwords.words('english'))
    
    # Add extra words to delete to the set of stopwords
    if remove:
        stop_words.update(remove)
    
    filtered_words = [word for word in all_words if word.isalnum() and word not in stop_words]
    word_counts = Counter(filtered_words)
    return word_counts

```

## Test an Instagram Account

```python
# account = 'leonardodicaprio'
# account='jenniferaniston'
# account='mattyperry4'
account='angelinajolie'
deleted_word= ['link', 'in', 'bio','lolavie','know','go']  # Add your extra words here
captions = get_captions(account)
word_counts = count_words(captions,deleted_word)
```

## plot a bar graph 
the code takes word counts, converts it into a DataFrame, and then visualizes the top N most repeated words in a bar graph.



```python
# Convert word counts to DataFrame
df_word_counts = pd.DataFrame.from_dict(word_counts, orient='index', columns=['count'])
df_word_counts.index.name = 'word'
df_word_counts.sort_values(by='count', ascending=False, inplace=True)

# Plot the top N most repeated words
top_n = 20
df_word_counts.head(top_n).plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title(f'Top {top_n} Most Repeated Words in Captions of {account}')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

# Sentiment Analysis on Captions
## TextBlob sentiment analysis

```python
import pandas as pd
from textblob import TextBlob

# Function to get sentiment
def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity, analysis.sentiment.subjectivity

# Apply the function to captions
df[['caption_polarity', 'caption_subjectivity']] = df['captions'].apply(lambda x: pd.Series(get_sentiment(str(x))))

# Display the dataframe with sentiment scores
df.head(10)
```

When using TextBlob for sentiment analysis, each text (in this case, the Instagram captions) is analyzed to produce two main scores: 
### 1.polarity
<br>
<br>


The polarity score ranges from -1.0 to 1.0.
-        -1.0: This indicates a very negative sentiment.
-        0.0: This indicates a neutral sentiment.
-        1.0: This indicates a very positive sentiment.

The polarity score helps you understand the overall sentiment expressed in the text. A higher positive score suggests positive sentiment, whereas a higher negative score suggests negative sentiment.
<br>
<br>
<br>


### 2.subjectivity.
<br>




The subjectivity score ranges from 0.0 to 1.0.
-       0.0: This indicates a very objective statement (fact-based).
-       1.0: This indicates a very subjective statement (opinion-based).

The subjectivity score helps you understand how much of the text is based on personal opinions or feelings versus objective facts. Higher scores indicate more subjective content.

<br>

By examining these scores in your dataframe, you can gain insights into the overall emotional tone and nature (subjective or objective) of your Instagram captions.

```python
def get_instagram_comments(username, password, target_profile, num_posts=2, num_comments=3):
    L = instaloader.Instaloader()
    L.login(username, password)
    profile = instaloader.Profile.from_username(L.context, target_profile)
    posts_data = []

    for post in profile.get_posts():
        if num_posts == 0:
            break
        num_posts -= 1
        comments = [comment.text for comment in post.get_comments()]
        limited_comments = comments[:num_comments]
        posts_data.append({
            'caption': post.caption,
            'comments': limited_comments,
            'likes': post.likes,
            'num_comments': len(limited_comments),
            'avg_comment_length': sum(len(comment) for comment in limited_comments) / len(limited_comments) if limited_comments else 0
        })

    return posts_data
```

## Sentiment Analysis on Comments

## Engagement Metrics
- Number of comments per post.
- Average length of comments.
- Like-to-comment ratio (if the number of likes is available from the post object).

```python
stop_words = set(stopwords.words('english'))

def clean_comment(comment):
    # Remove punctuation
    comment = comment.translate(str.maketrans('', '', string.punctuation))
    # Remove stopwords
    comment = ' '.join([word for word in comment.split() if word.lower() not in stop_words])
    return comment



def analyze_comments(posts_data):
    sentiment_data = []
    for post_data in posts_data:
        total_polarity = 0
        total_subjectivity = 0
        cleaned_comments = [clean_comment(comment) for comment in post_data['comments']]
        
        for comment in cleaned_comments:
            analysis = TextBlob(comment)
            total_polarity += analysis.sentiment.polarity
            total_subjectivity += analysis.sentiment.subjectivity
        
        avg_polarity = total_polarity / len(cleaned_comments) if cleaned_comments else 0
        avg_subjectivity = total_subjectivity / len(cleaned_comments) if cleaned_comments else 0
        
        sentiment_data.append({
            'caption': post_data['caption'],
            'average_polarity': avg_polarity,
            'average_subjectivity': avg_subjectivity,
            'likes': post_data['likes'],
            'num_comments': post_data['num_comments'],
            'avg_comment_length': post_data['avg_comment_length'],
            'like_to_comment_ratio': post_data['likes'] / post_data['num_comments'] if post_data['num_comments'] > 0 else 0
        })

    return sentiment_data

# Example usage
# posts_data = get_instagram_comments(username, password, target_profile, num_posts, num_comments)
sentiment_and_engagement_data = analyze_comments(posts_data)

# Print the results
for post_data in sentiment_and_engagement_data:
    print(f"Post: {post_data['caption']}")
    print(f"Average Polarity: {post_data['average_polarity']}")
    print(f"Average Subjectivity: {post_data['average_subjectivity']}")
    print(f"Likes: {post_data['likes']}")
    print(f"Number of Comments: {post_data['num_comments']}")
    print(f"Average Comment Length: {post_data['avg_comment_length']}")
    print(f"Like-to-Comment Ratio: {post_data['like_to_comment_ratio']}")
    print("\n")
```

1. Fetching Comments:
- The get_instagram_comments function logs into Instagram, fetches posts from the specified profile, and retrieves a limited number of comments for each post.
- It stores the caption, comments, number of likes, number of comments, and the average comment length for each post in posts_data.
2. Cleaning Comments:
- The clean_comment function removes punctuation and stopwords from each comment to prepare it for sentiment analysis.
3. Sentiment Analysis:
- The analyze_comments function processes each post's comments by cleaning them and then using TextBlob to analyze sentiment.
- It calculates the average polarity and subjectivity of comments for each post.
4. Engagement Metrics:
- The function also calculates engagement metrics for each post, including the number of comments, average comment length, and like-to-comment ratio.
5. Output:
- The script prints out the results for each post, including the sentiment analysis and engagement metrics.

```python
import re
import pandas as pd
from textblob import TextBlob

# Function to clean comments (assuming a simple placeholder here)
def clean_comment(comment):
    return comment

# Function to analyze comments and calculate sentiment
def analyze_comments(posts_data):
    sentiment_data = []
    all_cleaned_comments = []
    
    for post_data in posts_data:
        total_polarity = 0
        total_subjectivity = 0
        cleaned_comments = [clean_comment(comment) for comment in post_data['comments']]
        all_cleaned_comments.extend(cleaned_comments)
        
        for comment in cleaned_comments:
            analysis = TextBlob(comment)
            total_polarity += analysis.sentiment.polarity
            total_subjectivity += analysis.sentiment.subjectivity
        
        avg_polarity = total_polarity / len(cleaned_comments) if cleaned_comments else 0
        avg_subjectivity = total_subjectivity / len(cleaned_comments) if cleaned_comments else 0
        
        sentiment_data.append({
            'caption': post_data['caption'],
            'average_polarity': avg_polarity,
            'average_subjectivity': avg_subjectivity,
            'likes': post_data['likes'],
            'num_comments': post_data['num_comments'],
            'avg_comment_length': post_data['avg_comment_length'],
            'like_to_comment_ratio': post_data['likes'] / post_data['num_comments'] if post_data['num_comments'] > 0 else 0
        })

    return sentiment_data, all_cleaned_comments

# Function to extract hashtags from captions
def extract_hashtags(caption):
    return re.findall(r'#\w+', caption)

# Function to perform hashtag analysis
def hashtag_analysis(sentiment_data):
    hashtag_data = []

    for post_data in sentiment_data:
        hashtags = extract_hashtags(post_data['caption'])
        for hashtag in hashtags:
            hashtag_data.append({
                'hashtag': hashtag,
                'likes': post_data['likes'],
                'num_comments': post_data['num_comments'],
                'average_polarity': post_data['average_polarity'],
                'average_subjectivity': post_data['average_subjectivity']
            })

    return hashtag_data

# Function to correlate hashtags with engagement
def correlate_hashtags_engagement(hashtag_data):
    df = pd.DataFrame(hashtag_data)
    engagement_corr = df.groupby('hashtag').agg({
        'likes': 'mean',
        'num_comments': 'mean'
    }).reset_index()
    return engagement_corr

# Function to correlate hashtags with sentiment
def correlate_hashtags_sentiment(hashtag_data):
    df = pd.DataFrame(hashtag_data)
    sentiment_corr = df.groupby('hashtag').agg({
        'average_polarity': 'mean',
        'average_subjectivity': 'mean'
    }).reset_index()
    return sentiment_corr

# Analyze comments and calculate sentiment
sentiment_and_engagement_data, all_cleaned_comments = analyze_comments(posts_data)

# Perform hashtag analysis
hashtag_data = hashtag_analysis(sentiment_and_engagement_data)
engagement_corr = correlate_hashtags_engagement(hashtag_data)
sentiment_corr = correlate_hashtags_sentiment(hashtag_data)

# Display results
print("Hashtag Engagement Correlation:")
print(engagement_corr)

print("\nHashtag Sentiment Correlation:")
print(sentiment_corr)
```

## Word Cloud of Hashtags by Average Polarity

## Topic Modeling

```python
# Prepare the data for topic modeling
nlp = spacy.load('en_core_web_sm')
def preprocess(text):
    doc = nlp(text.lower())
    return ' '.join([token.lemma_ for token in doc if token.is_alpha and not token.is_stop])

# Preprocess comments and captions
# df['processed_comments'] = df['comments'].apply(lambda x: preprocess(str(x)))
df['processed_captions'] = df['captions'].apply(lambda x: preprocess(str(x)))

# Vectorize the text
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
# comments_vectorized = vectorizer.fit_transform(df['processed_comments'])
captions_vectorized = vectorizer.fit_transform(df['processed_captions'])

# Fit the LDA model
# lda_comments = LatentDirichletAllocation(n_components=5, random_state=42)
# lda_comments.fit(comments_vectorized)

lda_captions = LatentDirichletAllocation(n_components=5, random_state=42)
lda_captions.fit(captions_vectorized)

# Display the top words in each topic
def display_topics(model, feature_names, num_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[:-num_top_words - 1:-1]]))

num_top_words = 10
# print("Topics in comments:")
# display_topics(lda_comments, vectorizer.get_feature_names_out(), num_top_words)

print("Topics in captions:")
display_topics(lda_captions, vectorizer.get_feature_names_out(), num_top_words)
df.to_csv('captions1.csv', index=False)
df.head()
```
